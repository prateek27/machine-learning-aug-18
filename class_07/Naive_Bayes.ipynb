```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./mushrooms.csv')
df.head()
```

```python
le = LabelEncoder()
ds = df.apply(le.fit_transform)
```

```python
ds.head()
```

```python
data = ds.values
print(data)
```

```python
data_y = data[:,0]
data_x = data[:,1:]
print(data_x.shape)
print(data_y.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(data_x,data_y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
def prior_prob(x_train,y_train,label_value):
    n_rows = x_train[y_train==label_value]
    prob_val = n_rows.shape[0]/float(x_train.shape[0])
    return prob_val
prior_prob(x_train,y_train,1)
```

```python
def cond_prob(x_train,y_train,feature_col,feature_val,label_value):
    n_rows = x_train[y_train==label_value]
    constraint_rows = n_rows[n_rows[:,feature_col]==feature_val]
    prob_value = constraint_rows.shape[0]/float(n_rows.shape[0])
    return prob_value
```

```python
classes = np.unique(y_train)
print classes
```

```python
preds = []
for ix in range(x_test.shape[0]):
    post_prob = []
    for jx in classes:
        likelihood = 1.0
#         print(x_test[ix])
        for kx in x_test[ix]:
            cond = cond_prob(x_train,y_train,kx,x_test[ix,kx],jx)
            likelihood *= cond
        prior = prior_prob(x_train,y_train,jx)
        posterior_prob = likelihood*prior
#         print(posterior_prob)
        post_prob.append(posterior_prob)
    post_prob = np.array(post_prob)
#     print(post_prob)
    pred_label = post_prob.argmax()
    preds.append(pred_label)
```

```python
y_test = np.array(y_test)
preds = np.array(preds)
accuracy = np.sum(np.array(y_test==preds))/float(y_test.shape[0])
```
